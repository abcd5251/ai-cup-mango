```python
import cv2
import csv   #將每張圖片的bounding box 擷取出來並做分類 
import os    #做training的資料夾 
import numpy as np

base_dir= r"./training/" 
if not os.path.isdir(base_dir): os.mkdir(base_dir)  #製造training資料夾

#不良-機械傷害 :1  不良-炭疽病 :2  不良-著色不佳: 3  不良-乳汁吸附: 4 不良-黑斑病:5 
Train_dir= r"./Train/"
temp=0

for k in range(1,6):
    train_dir = os.path.join(base_dir, str(k))
    if not os.path.isdir(train_dir): os.mkdir(train_dir)
    

with open('train.csv', newline='',encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        print(temp)
        if temp==0:  #因為第一張圖片會有問題
            row[0]=row[0][1:]
        picture = os.path.join(Train_dir, str(row[0]))
        #print(len(row[0]))
        #print(picture)
        i=1
        #if temp==657: 
         #   print(len(row)) 這行最長會爆掉
        while row[i]!="": 
            x=int(row[i])
            y=int(row[i+1])
            w=int(row[i+2])
            h=int(row[i+3])  #選取bounding box
            what=str(row[i+4])
            i=i+5
         
            #print(x,y,w,h)
            if what=="不良-機械傷害":  #判斷種類並決定存在哪個資料夾  
                des=1
            elif what=="不良-炭疽病":
                des=2
            elif what=="不良-著色不佳":
                des=3
            elif what=="不良-乳汁吸附":
                des=4
            elif what=="不良-黑斑病":
                des=5
            if i>135: #因為每一行row裡面的值最多只會到第135個   不會再超過    
                break
            
            
            img = cv2.imread(picture)
            crop_img = img[y:y+h, x:x+w]  #找到特定範圍
            #cv2.imshow("cropped", crop_img)
            #cv2.waitKey(0)
            
            savepath= os.path.join(base_dir, str(des) ) #存入特定資料夾
            fileplace=os.path.join(savepath,str(row[0])) #存入的名稱
            if os.path.isfile(fileplace):  #isdir 是目錄 isfile是檔案
                print("iiii")
                fileplace=os.path.join(savepath,str(i)+str(row[0]))
                cv2.imwrite(fileplace, crop_img)   #考慮重複
            else:
                cv2.imwrite(fileplace, crop_img) 
            
        temp=temp+1
```

```python
#做 Validation 的資料夾
import cv2
import csv   #將每張圖片的bounding box 擷取出來並做分類 
import os    #做val的資料夾 
import numpy as np

temp=0

val_dir= r"./val/" 
.
validation_dir=r"./Dev/"
for k in range(1,6):
    dev_dir = os.path.join(val_dir, str(k))
    if not os.path.isdir(dev_dir): os.mkdir(dev_dir)
        
        
with open('dev.csv', newline='',encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        print(temp)
        if temp==0:  #因為第一張圖片會有問題
            row[0]=row[0][1:]
        picture = os.path.join(validation_dir, str(row[0]))
        #print(len(row[0]))
        #print(picture)
        i=1
        #if temp==657: 
         #   print(len(row)) 這行最長會爆掉
        while row[i]!="": 
            x=float(row[i])
            y=float(row[i+1])
            w=float(row[i+2])
            h=float(row[i+3])  #選取bounding box
            what=str(row[i+4])
            i=i+5
         
            #print(x,y,w,h)
            if what=="不良-機械傷害":  #判斷種類並決定存在哪個資料夾  
                des=1
            elif what=="不良-炭疽病":
                des=2
            elif what=="不良-著色不佳":
                des=3
            elif what=="不良-乳汁吸附":
                des=4
            elif what=="不良-黑斑病":
                des=5
            if i>135:
                break
            
            
            img = cv2.imread(picture)
            x=int(x)
            y=int(y)
            h=int(h)
            w=int(w)
            crop_img = img[y:y+h, x:x+w]  #找到特定範圍
            #cv2.imshow("cropped", crop_img)
            #cv2.waitKey(0)
            
            savepath= os.path.join(val_dir, str(des) ) #存入特定資料夾
            fileplace=os.path.join(savepath,str(row[0])) #存入的名稱
            if os.path.isfile(fileplace):  #isdir 是目錄 isfile是檔案
                print("iiii")
                fileplace=os.path.join(savepath,str(i)+str(row[0]))
                cv2.imwrite(fileplace, crop_img)   #考慮重複
            else:
                cv2.imwrite(fileplace, crop_img) 
            
        temp=temp+1
```

```python
#train 的檔案
#1 :502個檔案   2:23587個檔案  3:15045個檔案  4:2578個檔案  5: 1657個檔案

#43369個檔案

from tensorflow.keras import layers  # 從 keras 套件匯入 layers, models 套件
from tensorflow.keras import models
from tensorflow.keras import optimizers



model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150,150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))  #  在這裡加入 Dropout 層 (丟棄 50 %)
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))
model.summary()  # 查看模型摘要

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['acc'])
```

```python
#val 的檔案
#1 :91個檔案   2:3504個檔案  3:1997個檔案  4:391個檔案  5: 298個檔案
#6281個檔案
# 方法1   不做圖片的resize  用efficientNet 下去解
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping        
        
train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=40,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True ) #設定訓練、測試資料的 Python 產生器，並將圖片像素值依 1/255 比例重新壓縮到 [0, 1] 還有做資料的擴充

test_datagen = ImageDataGenerator(rescale=1./255)

height=150
width=150
batch_size=20

earlystop_callback = EarlyStopping(
  monitor='val_acc', min_delta=0.0001,
  patience=4) #耐心等多久  多久會變正常  單位epoch
```

```python
#os.remove(path)   #删除文件
#os.removedirs(path)   #删除空文件夹
 
#os.rmdir(path)    #删除空文件夹

import os 
import shutil
import random

val_dir= r"./val/" 


dir2= r"./training/2/" 
dir3= r"./training/3/" 
dir4= r"./training/4/" 
dir5= r"./training/5/" 

new_dir= r"./new_train/" 



file2_fnames = [os.path.join(dir2, fname) for
        fname in os.listdir(dir2)]
file3_fnames = [os.path.join(dir3, fname) for
        fname in os.listdir(dir3)]
file4_fnames = [os.path.join(dir4, fname) for
        fname in os.listdir(dir4)]
file5_fnames = [os.path.join(dir5, fname) for
        fname in os.listdir(dir5)]

for i in range(10):
    if i>=1:
        model=models.load_model('combine.h5')
    
    for k in range(2,6):
        neww = os.path.join(new_dir, str(k))
        if not os.path.isdir(neww): os.mkdir(neww)
    
    ff2=file2_fnames[520*i:520*(i+1)]
    ff3=file3_fnames[520*i:520*(i+1)]
    if i<=3:
        ff4=file4_fnames[520*i:520*(i+1)]
    if i >3:
        temp=i%4
        ff4=file4_fnames[520*temp:520*(temp+1)]
    if i<=2:
        ff5=file5_fnames[520*i:520*(i+1)]
    if i>2:
        temp=i%3
        ff5=file5_fnames[520*temp:520*(temp+1)]
  

    new_dir2=os.path.join(new_dir,"2")
    new_dir3=os.path.join(new_dir,"3")
    new_dir4=os.path.join(new_dir,"4")
    new_dir5=os.path.join(new_dir,"5")
        
    for fname in ff2:
        name=fname[13:]
        dst = os.path.join(new_dir2,name)
        shutil.copyfile(fname, dst)
    
    for fname in ff3:
        name=fname[13:]
        dst = os.path.join(new_dir3, name)
        shutil.copyfile(fname, dst)
      
    for fname in ff4:
        name=fname[13:]
        dst = os.path.join(new_dir4, name)
        shutil.copyfile(fname, dst)
    
    for fname in ff5:
        name=fname[13:]
        dst = os.path.join(new_dir5, name)
        shutil.copyfile(fname, dst)
    
    
    train_generator = train_datagen.flow_from_directory(
        new_dir,              # 目標目錄
        target_size=(height, width),  # 調整所有影像大小成 height,width
        batch_size=batch_size,
        class_mode='categorical')    


    validation_generator = test_datagen.flow_from_directory(
        val_dir,
        target_size=(height, width),
        batch_size=batch_size,
        class_mode='categorical')

    history = model.fit_generator(   
        train_generator,
        steps_per_epoch=int(2580/20),  
        epochs=15,
        callbacks=[earlystop_callback],
        validation_data=validation_generator,
        validation_steps=int(2000/20))  
       
    print("123")
    model=model.save('combine.h5')
   

    
    for k in range(2,6):
        neww = os.path.join(new_dir, str(k))
        if  os.path.isdir(neww): shutil.rmtree(neww)
            
```
