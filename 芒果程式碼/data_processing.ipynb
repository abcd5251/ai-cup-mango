```python
f = open('Test_mangoXYWH.csv', mode='r', encoding='UTF-8-sig') # 去除\ufeff
lines = f.readlines()
lines
```

```python
img_folder = './test'
label_folder = './all_label'
for line in lines[1:]:
    data = line.split(',')
    img_path = os.path.join(img_folder, data[0])
    img = cv2.imread(img_path)
    height = img.shape[0]
    width = img.shape[1]
    #print('height =', height, '; width =', width)
    
    # 定義x為width方向 y為height方向
    center_x = (float(data[1]) + float(data[3])/2) / width
    center_y = (float(data[2]) + float(data[4])/2) / height
    length_x = float(data[3]) / width
    length_y = float(data[4]) / height
    #print(center_x)
    #print(center_y)
    #print(length_x)
    #print(length_y)
    
    txt_path = os.path.join(label_folder, data[0].strip('.jpg') + '.txt')
    with open(txt_path,"w") as txt:   # 打開一個文件只用於寫入。如果該文件已存在則打開文件，並從開頭開始編輯，即原有內容會被刪除。如果該文件不存在，創建新文件。
        txt.write("0 %f %f %f %f\n" %(center_x, center_y, length_x, length_y))  # 這句話自帶檔案關閉功能，不需要再寫f.close()
```
