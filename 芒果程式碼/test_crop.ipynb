```python
import matplotlib.pyplot as plt
import pandas as pd
import cv2 as cv
import os
```

```python
df_test = pd.read_csv('Test_mangoXYWH.csv')
df_test.head(10)
```

```python
IMAGE = '23747.jpg'
plt.imshow(plt.imread('./test/' + IMAGE))
plt.show()
```

```python
for row in range(len(df_test)):
    x = df_test.loc[row, 'x']
    y = df_test.loc[row, 'y']
    w = df_test.loc[row, 'w']
    h = df_test.loc[row, 'h']
    img = cv.imread('./test/' + df_test.loc[row, 'image_id'])
    cv.imwrite('./test_crop/' + df_test.loc[row, 'image_id'], img[y:y+h, x:x+w])
```

```python
plt.imshow(plt.imread('./test_crop/' + IMAGE))
plt.show()
```

```python
import zipfile
# zipfile example
def zip_dir(path):
    zf = zipfile.ZipFile('{}.zip'.format(path), 'w', zipfile.ZIP_DEFLATED)
   
    for root, dirs, files in os.walk(path):
        for file_name in files:
            zf.write(os.path.join(root, file_name))
zip_dir('./yolov5')
```
