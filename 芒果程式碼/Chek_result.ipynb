```python
dirpath = r"./exp7/labels/"
result = [f for f in os.listdir(dirpath) if os.path.isfile(os.path.join(dirpath, f))]
```

```python
# print(result)
import re
Cord = []


for file in result:
    path = ""
    path += dirpath + file
    file = file[:-4]
    myfile = open(path,'r')
    lines = myfile.readlines()
    pattern = "class_name"
    Cord_Raw = []
    for line in lines:
        line = line[:-1]
        Cord_Raw = line.split(" ")
        Cord_Raw = Cord_Raw[1:]
        Cord_Raw.insert(0, file)
        Cord.append(Cord_Raw)
#         if re.search(pattern,line):
#             Cord_Raw = line
#             print(Cord_Raw)
#     Cord.append(Cord_Raw.split("(")[1].split(")")[0].split("  "))
```

```python
weird = set()
for i in range(len(Cord)):
    if i < len(Cord) - 1 and Cord[i][0] == Cord[i + 1][0]:
        weird.add(Cord[i][0])
```

```python
import cv2
dirpath = r"./exp7/"
save_dirpath = r"./exp7/"
timer = 0
for item in weird:
    count = 0
    for cord in Cord:
        if cord[0] == item:
            timer += 1
            count += 1
            path = ""
            path += dirpath + cord[0] + ".jpg"
            print(path)
            save_path = ""
            save_path += save_dirpath + cord[0] + "_" + str(count)+".jpg"
            img = cv2.imread(path)

            height, width, channels = img.shape
            
            x_min=int(width * float(cord[1]) - float(cord[3]) * width / 2)
            x_max=int(float(cord[3]) * width) + x_min

            y_min=int(height * float(cord[2]) - float(cord[4]) * height / 2)
            y_max=int(float(cord[4]) * height) + y_min

            crop_img = img[y_min:y_max, x_min:x_max]

            cv2.imwrite(save_path,crop_img)
print(timer)    
```

```python
import zipfile


# zipfile example
def zip_dir(path):
    zf = zipfile.ZipFile('{}.zip'.format(path), 'w', zipfile.ZIP_DEFLATED)
   
    for root, dirs, files in os.walk(path):
        for file_name in files:
            zf.write(os.path.join(root, file_name))


if __name__ == '__main__':
    path = './exp7'
    zip_dir(path)
```
